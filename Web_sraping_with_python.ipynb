```python
from bs4 import BeautifulSoup # ce module aide au web scraping.
import requests  # Ce module vous aide à téléchrger les pages web
```

```python
url = "https://en.wikipedia.org/wiki/Democracy_Index"
```

L'objet beatiful Soup transforme un documment HTML complex en une arborescence complexe.  

```python
# je vais stocker le contenu de la page
data = requests.get(url).text
```

```python
# Obet Beautiful soup
soup = BeautifulSoup(data, "html.parser")
```

```python
# Recupérons touts les liens
for link in soup.find_all('a', href = True) : # les liens sont représentés par <a>
 print(link.get('href'))
```

```python
# Scrapons toutes les images
for link in soup.find_all('img'): # dans Html, les images sont représentées par <img>
  print(link)
  print(link.get('src'))
```

## **Scrapons les données de l'objet html**

```python
# Charchons les tables html avec des données dans le html table
table = soup.find_all('table') # Dans html les tables sont représentées par <table>
```

```python
# Cherchons toutes les lignes de la table
for row in table: # Ici, table est un objet ResultSet contenant une liste d'éléments de table
    # Trouver toutes les colonnes de chaque ligne
    cols = row.find_all('td') # Dans HTML, les colonnes sont représentées par 'td'
    color_name = cols[2].string # Stocke la valeur dans la colonne 3 sous color_name
```
