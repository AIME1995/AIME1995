# **VISUALISATION DES DONNEES AVEC SEABORN EN UTILISANT NUMPY ET PANDAS**

#**Installation des packages nécessaires et création des données aléatoires**

```python
## Installons les packages nécessaires
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [6, 2]
```

```python
# Créons les données aléatoirement
dict = {
    #varibles qualitatives
  "genre": np.random.choice(["homme", "femme", "No-binaire"], 50),
  "niveau_education" : np.random.choice(["secondaires", "niversitaire","maîtrise", "Doctorat"], 50),
  "emploi": np.random.choice(["Technologie", "Finance", "santé", "Vente", "education"], 50),
  "satut_emploi": np.random.choice(["Temps plein", "Temps partiel", "Contractuel"], 50),
  "zone_residence": np.random.choice(["Urbain", "Rural", "Suburbain"], 50),
  "social" : np.random.choice(["Assurance maladie", "Régime de retraite", "Congés payés"], 50),
  'promotion' : np.random.choice(["elevees", "Moyennes", "Faibles"], 50),

    # variables quantitatives
  "age": np.random.randint(17 , 60, 50),
  "salaire": np.random.randint(1, 10, 50),
  "experience": np.random.randint(1, 30, 50),
  "Heures_semaine" : np.random.randint(6, 35, 50)
  }


df = pd.DataFrame(dict)
df.head()
```

# **Nuage de points (Scatter Plot)**

```python
sns.scatterplot(x = "experience", y ="salaire",
                data = df, hue = "genre")
```

#**Diagramme linéaire (Line Plot)**

```python
sns.lineplot(x="experience", y = "salaire", data = df)
```

# **Diagramme en barres (Bar Plot)**

```python
# diagramme en barres
sns.barplot(x="genre", y="salaire", data = df, hue = "genre")
```

# **Count Plot**

```python
sns.countplot( y = "salaire", data = df, hue = "zone_residence")
```

# **Histogramme (Histogram)**

```python
sns.histplot(x= "age", data=df)
```

# **Diagramme en violon (Violin Plot)**

```python
sns.violinplot(x = "niveau_education", y ="Heures_semaine", data = df, hue = "niveau_education")
```

#**Diagramme en boîte (Box Plot)**

```python
sns.boxplot(x = "emploi", y ="age", data = df, hue = "emploi")
```

```python
sns.boxenplot(x = "emploi", y ="age", data = df, hue = "emploi")
```

#**Heatmap**

```python
sns.heatmap(df[['age', "salaire",	"experience",	"Heures_semaine"]].corr())
```

#**stripplot**

```python
sns.stripplot(x = "emploi", y ="age", data = df, hue = "emploi")
```

#**Carte de cluster**

```python
sns.clustermap(df.drop('salaire', axis = 1))
```

# **Pairplot**

```python
sns.pairplot(df, hue = "genre", height = 2)
plt.savefig("paire.png")
```
